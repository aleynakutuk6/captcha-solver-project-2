```python
import random
import string
import cv2
from PIL import Image
from claptcha.claptcha import Claptcha
from os import listdir
from os.path import isfile, join
from random import randrange

def randomString():
    rndLetters = (random.choice(string.ascii_uppercase) for _ in range(1))
    return "".join(rndLetters)

def myString(i):
    return string.digits[i]

dir = "fonts"
out_dir = "outputs3"
files = [f for f in listdir(dir) if isfile(join(dir, f))]
for filename in files:
  for i in range(0,10):
    for j in range(0,10):
      rs = randrange(3)
      if rs == 0:
        resample=Image.BICUBIC
      elif rs == 1:
        resample=Image.BILINEAR
      elif rs == 2:
        resample=Image.NEAREST
      c = Claptcha(myString(i), dir+"/"+filename, (60,100), resample=resample, noise=random.uniform(0, 1))
      filepath = out_dir+"/"+str(i)+"_"+filename+"_"+str(j)+".png"
      text, _ = c.write(filepath)

```
